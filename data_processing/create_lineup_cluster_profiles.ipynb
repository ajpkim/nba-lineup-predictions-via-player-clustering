```python

def get_lineup_cluster_profile(lineup, df_players):
    """
    Function for building lineup cluster profiles (bpm weighted and non bpm weighted)
    based on player ids and year using a dataset with player cluster assignments and bpm weights.
    
    Return a 1 row dataframe with same index as lineup and columns for cluster profiles.
    """
    
#     print('Called with lineup:', lineup)
    
    player_ids = lineup['player_ids'].split(', ')
    cluster_bpm_cols = [f'cluster_{n}_bpm' for n in range(12)]
    cluster_nonbpm_cols = [f'cluster_{n}' for n in range(12)]    
    df_lineup_clusters = pd.DataFrame({col: 0.0 for col in cluster_bpm_cols+cluster_nonbpm_cols}, index=[lineup.name])
    print('all lineup pids:', player_ids)
    print('year:', year)
    for pid in player_ids:
        player = df_players[(df_players['year'] == year) & (df_players['player_id'] == pid)]
        print('player: ', pid)
        df_lineup_clusters[cluster_bpm_cols] += player[cluster_bpm_cols].to_numpy().reshape(-1)
        df_lineup_clusters[cluster_nonbpm_cols] += player[cluster_nonbpm_cols].to_numpy().reshape(-1)
    
#     print('returning:', df_lineup_clusters)
    print(df_lineup_clusters.shape)
    print('---------------------')
    return df_lineup_clusters
```
